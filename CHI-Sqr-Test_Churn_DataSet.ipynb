```python
import pandas as pd
```

```python
ch = pd.read_csv('Churn.csv')
```

```python
ch.head(3)
```

```python
ch.drop(['customerID'],axis =1,inplace=True)  # drop column which is not required
```

```python
y = ch.drop(['Churn'],axis =1)
```

```python
from scipy.stats import chi2_contingency
```

```python
chi_sqr_values = []
for i in y:
    c = pd.crosstab(ch['Churn'],ch[i])
    Pvalue = chi2_contingency(c)[1]
    chi_sqr_values.append(Pvalue)
```

```python
chi_sqr_values
```

```python
df = pd.DataFrame()
```

```python
ch.columns[0:-1]
```

```python
list(ch.columns)[0:-1]
```

```python
df['columns'] = list(y)
df['Pvalues'] = chi_sqr_values
```

```python
df
```

```python
df = df.sort_values(['Pvalues'])
```

```python
df[df.Pvalues < .05]
```
