```python
# take mpg as your target variavble and find corelation  of mpg with all variables
# create a df which has 2 columns 'nam', co-relation
```

```python
names =list(mtcars.columns)
```

```python
names.remove('mpg')
names
```

```python
corr_list = []
for i in names:
    corr_list.append(mtcars.mpg.corr(mtcars[i]))
```

```python
corr_list 
```

```python
corr_df = pd.DataFrame()
```

```python
corr_df ['col_names'] = names
corr_df['Co-relation'] = corr_list
```

```python
corr_df
```

```python
corr_df.sort_values('Co-relation',ascending= False)
```
